# **University Course Management System Project**
This Course Management System models various courses offered at COMSATS University, using core Object-Oriented Programming (OOP) principles. The system handles different types of courses (Programming, Mathematics, Science, Language) and their unique requirements, while also managing student enrollments.

The project demonstrates the following OOP principles:
# **1. Inheritance**
A base class course serves as a general template for all types of courses, containing shared attributes and methods.
Specialized course types (programming_course, mathematics_course, science_course, and language_course) inherit from the base class, reusing and extending its properties.
# **2. Encapsulation**
The class uses private attributes (e.g., __Course_name, __Credits) to restrict direct access and enforce data integrity.
Getter methods (like get_Course_name() and get_Credits()) provide controlled access to these private attributes.
# **3. Polymorphism**
The method get_schedule() is overridden in the child classes to provide course-specific schedules, demonstrating polymorphism.
The course_info() function dynamically adjusts the output based on the type of course passed to it (e.g., it shows practical exercises for programming courses but lab manuals for science courses).
# **4. Abstraction**
The complex details of enrolling students are hidden within the enroll_student() function, and users only need to provide the studentâ€™s name.
Similarly, the get_syllabus() method hides internal data handling and simply provides syllabus information.

# **Code Design and Flow**
**1. Base Class:**

*   The course class defines common attributes for all courses (course name,credits, instructor, etc.) and methods such as get_syllabus() and get_schedule().

**Derived Classes:**
*   programming_course adds a programming_language attribute and methods for practical exercises.
*   mathematics_course introduces a math_field attribute and formula sheets.
science_course adds a science_field attribute and lab manuals.
*   language_course manages language-specific materials with grammar guides.

**Enrollment Management:**
*   Each course maintains a list of enrolled students.
*   The enroll_student() function allows dynamic student registration and updates the course-specific student list.

**Course Information Display:**
*   The course_info() function provides detailed information for each course, including schedules, enrolled students, and course-specific resources.


```python
# Base class 'course' to represent a generic course
class course:
    def __init__(self, Course_name, Credits, Course_code, Instructor, Semester):
        # Private attributes to ensure Encapsulation (data hiding)
        self.__Course_name = Course_name
        self.__Credits = Credits
        self.__Course_code = Course_code
        self.__Instructor = Instructor
        self.__Semester = Semester

    # Method to get the syllabus for the course (Abstraction: hides internal data management)
    def get_syllabus(self):
        return f'Syllabus for {self.__Course_name}'

    # Method to get the course name (used for display purposes)
    def get_Course_name(self):
        return self.__Course_name

    # Getter methods to retrieve private course details (Encapsulation through getters)
    def get_Credits(self):
        return self.__Credits

    def get_Course_code(self):
        return self.__Course_code

    def get_Instructor(self):
        return self.__Instructor

    def get_Semester(self):
        return self.__Semester

    # Method to get the course schedule (can be overridden in subclasses to achieve Polymorphism)
    def get_schedule(self):
        return f"{self.__Course_name} is scheduled for 3 lectures per week."


# Specialized course classes inheriting from 'course' (Inheritance)
class programming_course(course):
    def __init__(self, Course_name, Credits, Course_code, Instructor, Semester, programming_language):
        # Inherit properties from the base class using 'super()' (Inheritance)
        super().__init__(Course_name, Credits, Course_code, Instructor, Semester)
        self.programming_language = programming_language

    # Method specific to programming courses
    def get_practical_exercises(self):
        return f'Practice exercises will involve coding in {self.programming_language}'

    # Overriding the base class method to achieve Polymorphism
    def get_schedule(self):
        return f"{self.get_Course_name()} includes coding assignments and 1 lab session per week."


class mathematics_course(course):
    def __init__(self, Course_name, Credits, Course_code, Instructor, Semester, math_field):
        super().__init__(Course_name, Credits, Course_code, Instructor, Semester)
        self.math_field = math_field

    def get_formula_sheet(self):
        return f"The course covers important formulas in {self.math_field}."

    def get_schedule(self):
        return f"{self.get_Course_name()} includes 4 theory lectures and 1 problem-solving session."


class science_course(course):
    def __init__(self, Course_name, Credits, Course_code, Instructor, Semester, science_field):
        super().__init__(Course_name, Credits, Course_code, Instructor, Semester)
        self.science_field = science_field

    def get_lab_manual(self):
        return f"Lab manual for {self.science_field} is available."

    def get_schedule(self):
        return f"{self.get_Course_name()} includes 4 lectures and 1 lab experiment per week."


class language_course(course):
    def __init__(self, Course_name, Credits, Course_code, Instructor, Semester, language):
        super().__init__(Course_name, Credits, Course_code, Instructor, Semester)
        self.language = language

    def get_language_guide(self):
        return f"A guide to {self.language} grammar and vocabulary is available."

    def get_schedule(self):
        return f"{self.get_Course_name()} includes language practice and oral assessments."


# Function to enroll new students and update the existing student list
def enroll_student(course, student_list):
    print(f"\nCurrently enrolled students in {course.get_Course_name()}: {', '.join(student_list)}")
    new_student = input("Enter the name of the student to enroll: ").strip()  # Get input from user

    if new_student:  # If the user provides a valid name
        student_list.append(new_student)  # Add the student to the list (Encapsulation: list management)
        print(f"{new_student} has been successfully enrolled in {course.get_Course_name()}!")
    else:
        print("No name entered. Enrollment canceled.")

    # Show updated student list after enrollment
    print(f"Updated student list for {course.get_Course_name()}: {', '.join(student_list)}\n")

    return student_list  # Return the updated list of students


# Function to display course details along with enrolled students (Abstraction)
def course_info(course, student_names):
    print('    ********************\n   ')
    print(f"Course Name: {course.get_Course_name()}")
    print(f"Credits: {course.get_Credits()}")
    print(f"Course Code: {course.get_Course_code()}")
    print(f"Instructor: {course.get_Instructor()}")
    print(f'Enrolled Students: {", ".join(student_names)}')
    print(f"Semester: {course.get_Semester()}")
    print(f"Syllabus: {course.get_syllabus()}")
    print(f"Schedule: {course.get_schedule()}")

    # Display specific details for different course types (Polymorphism)
    if isinstance(course, programming_course):
        print(f"Practical Exercises: {course.get_practical_exercises()}")
    elif isinstance(course, mathematics_course):
        print(f"Formula Sheet: {course.get_formula_sheet()}")
    elif isinstance(course, science_course):
        print(f"Lab Manual: {course.get_lab_manual()}")
    elif isinstance(course, language_course):
        print(f"Language Guide: {course.get_language_guide()}")


# Creating course objects (Inheritance: creating specialized courses from the base class)
course1 = programming_course('Programming Fundamentals', 3, 'FM-409', 'Dr. Najeeb', 3, 'Python')
course2 = mathematics_course('Discrete Mathematics', 3, 'CM-304', 'Dr. Fouzia', 4, 'Set Theory')
course3 = science_course('Physics', 3, 'PHY-101', 'Dr. Tanveer', 1, 'Mechanics')
course4 = language_course('English', 3, 'ENG-102', 'Miss Anum', 2, 'English')

# Initial student lists for each course
students_course1 = ['Tauseef', 'Rehman', 'Dania']
students_course2 = ['Minhal', 'Abdullah', 'Tanzila']
students_course3 = ['Zain', 'Fizza', 'Zainab']
students_course4 = ['Mujeeb', 'Sarim', 'Awais']

# Displaying course info and enrolling new students (Polymorphism in course-specific details)
course_info(course1, students_course1)
students_course1 = enroll_student(course1, students_course1)  # Update the list with new enrollment

course_info(course2, students_course2)
students_course2 = enroll_student(course2, students_course2)  # Update the list with new enrollment

course_info(course3, students_course3)
students_course3 = enroll_student(course3, students_course3)  # Update the list with new enrollment

course_info(course4, students_course4)
students_course4 = enroll_student(course4, students_course4)  # Update the list with new enrollment
```
